```python
import sqlalchemy as sal

engine = sal.create_engine('mssql://writeyourSQLDashboardName/master?driver=writeYourSQLODBCDRIVER')
```

```python
# Create a new database
import logging

logger = logging.getLogger(__name__)
new_db = 'walmart_db'

try:
    with engine.connect() as conn:
        conn.execution_options(isolation_level="AUTOCOMMIT")
        conn.execute(sal.text(f"CREATE DATABASE {new_db}"))
        logger.info("Data base has been created successful.")
except Exception as e:
    logger.error(f"Error occured while creating the database in SQL server data {e}")
    raise
```

```python
# Connect to your database

server = 'writeyourSQLDashboardName'
database = 'walmart_db'
driver = 'writeYourSQLODBCDRIVER'

engine = sal.create_engine(
    f"mssql+pyodbc://@{server}/{database}"
    f"?driver={driver}&trusted_connection=yes"
)

conn = engine.connect()
```

```python
logging.basicConfig(
    level=logging.INFO,
    force=True
)
try:
    df.to_sql(
        'df_walmart',
        con=conn,
        index=False,
        if_exists='replace'
    )
    logger.info("Table has been created successfully in SQL database.")
except Exception as e:
    logger.error(f"Error occured while creating the table inside the database in SQL server data {e}")
    raise
```
