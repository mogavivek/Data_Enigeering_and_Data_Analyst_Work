Also handling the foreihn characters with the use of NVARCHAR

```python
from sqlalchemy.types import NVARCHAR

df.to_sql(
    "df_netflix_raw",
    engine,
    if_exists="append",
    index=False,
    dtype={
        "title": NVARCHAR(200),
        "director": NVARCHAR(250),
        "cast": NVARCHAR(1000),
        "country": NVARCHAR(150),
        "description": NVARCHAR(500),
    }
)
```
